Using sendy_logistics.csv extract user_id in a seperate column.

```python
import seaborn as sns
import pandas as pd
import numpy as np

sns.set(rc={'figure.figsize':[20,15]},font_scale=1.1)
```

```python
df=pd.read_csv('sendy_logistics.csv')
df.head()
```

```python
def id(x):
    return x.split('_')[2]

df['ID']=df['User Id'].apply(id)

df.head()
```
