extract Mr,Miss,Mrs,etc.. from 'Name' column and use it to create new column and call it 'Title', then plot a countplot 

```python
import seaborn as sns
import pandas as pd
import numpy as np

sns.set(rc={'figure.figsize':[20,15]},font_scale=1.1)
```

```python
df=pd.read_csv('titanic_full.csv')
df.head()
```

```python
def name(x):
    x=x.split(',')[1]
    return x.split('.')[0]

df['Title']=df['Name'].apply(name)
```

```python
df
```

```python
sns.countplot(x='Title',data=df)
```
